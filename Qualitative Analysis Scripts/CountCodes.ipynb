### Counter Initialization

```python
# initialize counters by code (replacing - with _)
COM_PSR = 0
COM_EFF = 0
COM_CLR = 0

CLB_COP = 0
CLB_MNG = 0
CLB_DEL = 0
CLB_CMR = 0

PRO_STR = 0
PRO_ADP = 0
PRO_GOL = 0

INT_GRP = 0
INT_OTH = 0
INT_ENC = 0
INT_PAT = 0
INT_PER = 0
INT_BND = 0
INT_AWR = 0
INT_VUL = 0

SEF_DTR = 0
SEF_LCK = 0
SEF_FLW = 0

DEF_SLO = 0
DEF_IRR = 0
DEF_ISO = 0

CTR_DCT = 0
CTR_LWF = 0

DEV_NRL = 0
DEV_VCM = 0
```

### Function Definitions (reset counters, incrementor, iterator)

```python
# reset all counters to 0
def reset_counts(COM_PSR, COM_EFF, COM_CLR, CLB_COP, CLB_MNG, CLB_DEL, CLB_CMR, PRO_STR, PRO_ADP, PRO_GOL, INT_GRP, INT_OTH, INT_ENC, INT_PAT, INT_PER, INT_BND, INT_AWR, INT_VUL, SEF_DTR, SEF_LCK, SEF_FLW, DEF_SLO, DEF_IRR, DEF_ISO, CTR_DCT, CTR_LWF, DEV_NRL, DEV_VCM):
    COM_PSR = 0
    COM_EFF = 0
    COM_CLR = 0

    CLB_COP = 0
    CLB_MNG = 0
    CLB_DEL = 0
    CLB_CMR = 0

    PRO_STR = 0
    PRO_ADP = 0
    PRO_GOL = 0

    INT_GRP = 0
    INT_OTH = 0
    INT_ENC = 0
    INT_PAT = 0
    INT_PER = 0
    INT_BND = 0
    INT_AWR = 0
    INT_VUL = 0

    SEF_DTR = 0
    SEF_LCK = 0
    SEF_FLW = 0

    DEF_SLO = 0
    DEF_IRR = 0
    DEF_ISO = 0

    CTR_DCT = 0
    CTR_LWF = 0

    DEV_NRL = 0
    DEV_VCM = 0
```

```python
def increment_corr_code(row):
    global COM_PSR, COM_EFF, COM_CLR, CLB_COP, CLB_MNG, CLB_DEL, CLB_CMR, PRO_STR, PRO_ADP, PRO_GOL, INT_GRP, INT_OTH, INT_ENC, INT_PAT, INT_PER, INT_BND, INT_AWR, INT_VUL, SEF_DTR, SEF_LCK, SEF_FLW, DEF_SLO, DEF_IRR, DEF_ISO, CTR_DCT, CTR_LWF, DEV_NRL, DEV_VCM
    if(row['CODE'] == "COM-PSR"):
        COM_PSR += 1
    if(row["CODE"] == "COM-EFF"):
        COM_EFF += 1
    if(row["CODE"] == "COM-CLR"):
        COM_CLR += 1
    if(row["CODE"] == "CLB-COP"):
        CLB_COP += 1
    if(row["CODE"] == "CLB-MNG"):
        CLB_MNG += 1
    if(row["CODE"] == "CLB-DEL"):
        CLB_DEL += 1
    if(row["CODE"] == "CLB-CMR"):
        CLB_CMR += 1
    if(row["CODE"] == "PRO-STR"):
        PRO_STR += 1
    if(row["CODE"] == "PRO-ADP"):
        PRO_ADP += 1
    if(row["CODE"] == "PRO-GOL"):
        PRO_GOL += 1
    if(row["CODE"] == "INT-GRP"):
        INT_GRP += 1
    if(row["CODE"] == "INT-OTH"):
        INT_OTH += 1
    if(row["CODE"] == "INT-ENC"):
        INT_ENC += 1
    if(row["CODE"] == "INT-PAT"):
        INT_PAT += 1
    if(row["CODE"] == "INT-PER"):
        INT_PER += 1
    if(row["CODE"] == "INT-BND"):
        INT_BND += 1
    if(row["CODE"] == "INT-AWR"):
        INT_AWR += 1
    if(row["CODE"] == "INT-VUL"):
        INT_VUL += 1
    if(row["CODE"] == "SEF-DTR"):
        SEF_DTR += 1
    if(row["CODE"] == "SEF-LCK"):
        SEF_LCK += 1
    if(row["CODE"] == "SEF-FLW"):
        SEF_FLW += 1
    if(row["CODE"] == "DEF-SLO"):
        DEF_SLO += 1
    if(row["CODE"] == "DEF-IRR"):
        DEF_IRR += 1
    if(row["CODE"] == "DEF-ISO"):
        DEF_ISO += 1
    if(row["CODE"] == "CTR-DCT"):
        CTR_DCT += 1
    if(row["CODE"] == "CTR-LWF"):
        CTR_LWF += 1
    if(row["CODE"] == "DEV-NRL"):
        DEV_NRL += 1
    if(row["CODE"] == "DEV-VCM"):
        DEV_VCM += 1
```

```python
# iterate through every row of the passed dataframe and call the incrementing func.
def code_counter(df):
    for index, row in df.iterrows():
        increment_corr_code(row)
```

### Results Table (create, print, save)

```python
def results_table(df_low, df_high):
    # retrieve counts for both groups
    low_counts = df_low['CODE'].value_counts().reset_index()
    high_counts = df_high['CODE'].value_counts().reset_index()
    
    # merge into one table
    comparison = pd.merge(low_counts, high_counts, on='CODE', how='outer', suffixes=('_Low', '_High')).fillna(0)
    comparison.columns = ['Code', 'Low_Freq', 'High_Freq']
    
    # split codes (retrieves prefix category) and add to new column called CATEGORY
    comparison['CATEGORY'] = comparison['Code'].str.split('-').str[0]
    
    # category totals
    totals = comparison.groupby('CATEGORY')[['Low_Freq', 'High_Freq']].sum().reset_index()
    totals['Code'] = 'TOTAL'
    
    # combine and sort (push ***TOTAL*** to the bottom of each group)
    master = pd.concat([comparison, totals])
    master['Is_Total'] = master['Code'] == 'TOTAL'
    master = master.sort_values(['CATEGORY', 'Is_Total', 'Code'])
    
    # add low frequency and high frequency counts to calculate counts for the combined datframe 
    master['Combined'] = master['Low_Freq'] + master['High_Freq']
    
    # select columns and convert to integers rather than floats (looks cleaner)
    master = master[['CATEGORY', 'Code', 'Low_Freq', 'High_Freq', 'Combined']].reset_index(drop=True)
    master[['Low_Freq', 'High_Freq', 'Combined']] = master[['Low_Freq', 'High_Freq', 'Combined']].astype(int)

    # bold text and add shading for a more readable table
    styled_df = master.style.hide(axis='index') \
            .set_properties(**{'text-align': 'center', 'width': '120px', 'border': '1px solid #eeeeee'}) \
            .set_table_styles([
                {'selector': 'th', 'props': [('text-align', 'center'), ('font-weight', 'bold'), ('background-color', '#f2f2f2')]}
            ]) \
            .apply(lambda row: ['font-weight: bold; background-color: #f8f9fa' if row['Code'] == 'TOTAL' else '' for _ in row], axis=1)
    
    return styled_df
        
results_table = results_table(df_low_freq, df_high_freq)
display(results_table)

# try catch to safely save results table as a CSV to specific given file path 
try:
    results_table.data.to_csv('Gaming_Survey_Code_Counts.csv', index=False)
    print(f"Successfully saved total code counts to Gaming_Survey_Code_Counts.csv in folder: '{folder_name}'")
except Exception as e:
    print(f"AN UNEXPECTED ERROR OCCURRED: {e}")
```
