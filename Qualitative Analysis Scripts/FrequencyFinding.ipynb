# Frequency Finding Script for CS580B2 Group Project

### Imports

```python
import pandas as pd
import numpy as np
from IPython.display import display
import os
```

### Load CSVs as DataFrames

```python
df_FA25 = pd.read_csv('gaming25fa.csv', index_col=False, encoding='latin1')
print("Fall 2025 data successfully loaded.")
print(df_FA25.shape)

df_SP25 = pd.read_csv('gaming25sp.csv', index_col=False, encoding='latin1')
print("Spring 2025 data successfully loaded.")
print(df_SP25.shape)

df_combined = pd.concat([df_SP25, df_FA25], ignore_index=True)
print(df_combined.shape)
```

### Clean DatafFrames

```python
# replace any blank cells with NaN
df_combined = df_combined.replace(r'^\s*$', np.nan, regex=True)

# clean dataframe of any rows that may be NaN (precaution)
df_combined = df_combined.dropna()
```

### Verify Number of Rows

```python
print(f"Spring rows: {len(df_SP25)}")
print(f"Fall rows: {len(df_FA25)}")
print(f"Combined rows: {len(df_combined)}")
```

### Create Empty Dataframes for Low and High Frequency Gamers

```python
# define column names
column_names = ['final score', '7017232: I am a gamer.', '7017233: How many days per week do you game?', '7017234: How often do you game in a team?', '7017235: List the top 5 games you play with the most frequent first.', '7017236: Does your gaming make you a better team player? How?', 'CODE']

# create empty dataframes with specified column names
df_low_freq = pd.DataFrame(columns=column_names)
df_high_freq = pd.DataFrame(columns=column_names)

# extract low frequency (0-3 days) gamers from combined df
df_low_freq = df_combined[df_combined['7017233: How many days per week do you game?'].isin([0, 1, 2, 3])]

# extract high frequency (4-7 days) gamers from combined df
df_high_freq = df_combined[df_combined['7017233: How many days per week do you game?'].isin([4, 5, 6, 7])]
```

### Save Master DataFrame (Combined SP+FA), Low Frequency DataFrame, and High Frequency DataFrame

```python
# try catch to safely save each dataframe as a CSV to specific given file path 
folder_name = "Downloads"
try:
    df_combined.to_csv('Gaming_Survey_Combined.csv')
    print(f"Successfully saved dataframe containing all rows with student responses to Gaming_Survey_Combined.csv saved in folder: '{folder_name}'")

    df_low_freq.to_csv('Gaming_Survey_Low_Freq.csv', index=False)
    print(f"Successfully saved dataframe containing all rows with students who game 0-3 days/wk to Gaming_Survey_Low_Freq.csv in folder: '{folder_name}'")

    df_high_freq.to_csv('Gaming_Survey_High_Freq.csv', index=False)
    print(f"Successfully saved dataframe containing all rows with students who game 4-7 days/wk to Gaming_Survey_High_Freq.csv in folder: '{folder_name}'")
except Exception as e:
    print(f"AN UNEXPECTED ERROR OCCURRED {e}")

print(df_low_freq.shape)
print(df_high_freq.shape)
```
