### Imports

```python
import pandas as pd
import os
```

## Assign Random Number Generator Seed, Percentage of Rows to Audit, and Directory

```python
random_seed = 42
audit_percentage = 0.25
directory_path = r"LLM-Coded-Data"
```

### Run Audit Report, Print Results

```python
audit_report = {}
for filename in os.listdir(directory_path):
    if filename.endswith(".csv"):
        file_path = os.path.join(directory_path, filename)
        df = pd.read_csv(file_path)

        # calculate sample size for this specific cohort
        total_rows = len(df)
        sample_size = int(total_rows * audit_percentage)

        # generate random indices using a fixed seed
        audit_indices = df.sample(n=sample_size, random_state=random_seed).index.tolist()
            
        # store results
        audit_report[filename] = {
            "Total Responses": total_rows,
            "Audit Count": sample_size,
            "Indices to Verify": sorted(audit_indices)
            }

        print(f"Cohort: {filename}")
        print(f"Total Rows: {total_rows}")
        print(f"Audit Rows (25%): {sample_size}")
        print(f"Indices: {sorted(audit_indices)}\n")
```
